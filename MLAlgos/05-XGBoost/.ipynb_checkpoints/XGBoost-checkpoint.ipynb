1. Import required libraries

```python
import pandas as pd
import os
from sklearn.pipeline import Pipeline
from category_encoders.target_encoder import TargetEncoder
from xgboost import XGBClassifier
#pip install scikit-optimize
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer
from xgboost import plot_importance
```

2. Get train data

```python
selected_cols = ["Client_Income", "Car_Owned", "Bike_Owned", "Active_Loan", "House_Own", "Child_Count"]
selected_cols_ex = ["Credit_Amount", "Loan_Annuity", "Client_Income_Type", "Client_Education"]
selected_cols.extend(selected_cols_ex)
selected_cols_ex = ["Client_Marital_Status", "Client_Gender", "Loan_Contract_Type", "Client_Housing_Type"]
selected_cols.extend(selected_cols_ex)
selected_cols_ex = ["Age_Days", "Client_Occupation", "Client_Family_Members"]
selected_cols.extend(selected_cols_ex)
selected_cols_ex = ["Score_Source_1", "Score_Source_2", "Score_Source_3", "Default"]
selected_cols.extend(selected_cols_ex)
```

```python
df_train = pd.read_csv (os.path.join("../../Data/NBFI_Vehicle_Loan_Defaults", "Train_Dataset.csv"), usecols=selected_cols, low_memory=False)
```

```python
df_train.info()
```

```python
#Check if any columns gets an undesired type
df_train["Score_Source_3"] = pd.to_numeric(df_train["Score_Source_3"], errors="coerce")
df_train["Loan_Annuity"] = pd.to_numeric(df_train["Loan_Annuity"], errors="coerce")
df_train["Age_Days"] = pd.to_numeric(df_train["Loan_Annuity"], errors="coerce")
```

```python
# Clean columns
df_train = df_train.dropna(subset=["Client_Income"] )
df_train = df_train.dropna(subset=["Credit_Amount"])
df_train = df_train.dropna(subset=["Loan_Annuity"])
df_train["Car_Owned"] = df_train["Car_Owned"].fillna(0)
df_train["Bike_Owned"]= df_train["Bike_Owned"].fillna(0)
df_train["Active_Loan"] = df_train["Active_Loan"].fillna(0)
df_train["House_Own"] = df_train["House_Own"].fillna(0)
df_train["Child_Count"] = df_train["Child_Count"].fillna(df_train["Child_Count"].median())
df_train["Client_Income_Type"] = df_train["Client_Income_Type"].fillna("Unknown")
df_train["Client_Education"] = df_train["Client_Education"].fillna("Unknown")
df_train["Client_Marital_Status"] = df_train["Client_Marital_Status"].fillna("U")
df_train["Client_Gender"] = df_train["Client_Gender"].fillna("XNA")
df_train["Loan_Contract_Type"] = df_train["Loan_Contract_Type"].fillna("CL")
df_train["Client_Housing_Type"] = df_train["Client_Housing_Type"].fillna("Unknown")
df_train["Client_Occupation"] = df_train["Client_Occupation"].fillna('Unknown')
df_train["Client_Family_Members"] = df_train["Client_Family_Members"].fillna(df_train["Client_Family_Members"].median())
df_train["Score_Source_1"] = df_train["Score_Source_1"].fillna(df_train["Score_Source_1"].mean())
df_train["Score_Source_2"] = df_train["Score_Source_2"].fillna(df_train["Score_Source_2"].mean())
df_train["Score_Source_3"] = df_train["Score_Source_3"].fillna(df_train["Score_Source_3"].mean())
```

```python
print(df_train.info())
```

```python
y_train = df_train["Default"]
```

```python
y_train.info()
```

```python
df_train = df_train.drop(["Default"], axis='columns')
```

3.TargetEncoders - convert categorical variables into numeric

```python
estimators = [ ('encoder', TargetEncoder()), ('clf', XGBClassifier(random_state = 2025))  ]
```

```python
pipe = Pipeline(steps=estimators)   
print(pipe)
```

4.Set Hyper parameters

```python
search_space = {
        'clf__max_depth': Integer(2,8),
        'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
        'clf__subsample': Real(0.5, 1.0),
        'clf__colsample_bytree' : Real(0.5, 1.0),
        'clf__colsample_bylevel' : Real(0.5, 1.0),
        'clf__colsample_bynode' : Real(0.5, 1.0),
        'clf__reg_alpha' : Real(0.0, 10),
        'clf__reg_lamda' : Real(0.0, 10),
        'clf__gamma' : Real(0.0, 10)
    }
```

```python
opt = BayesSearchCV(pipe, search_space, cv=3, n_iter=10, scoring='roc_auc', random_state=2025)
```

5. Train XGBoost by callign Fit

```python
opt.fit(df_train, y_train)
```

6. Evluate the model

```python
opt.best_estimator_
```

```python
opt.best_score_
```

7. Meausre feature importance

```python
opt.best_estimator_.steps
```

```python
xgboost_step = opt.best_estimator_.steps[1]
xgboost_model = xgboost_step[1]
plot_importance(xgboost_model)
```
